**IMPORTING LIBRARIES**

```python
import pandas as pd
import ast 
```

<br>

**LOAD DATASET**

```python
# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv")
df.head()
```

<br>

**COLUMNS SELECTION FOR ANALYSIS**

```python
# Select only the necessary columns
df = df[['title', 'genres']]

# Display the first few rows to verify the data
df.head()
```

<br>

**PROCESS THE GENRES**

```python
def extract_genres(genre_json):
    """
    Extract the genre names from a JSON string.
    """
    try:
        genres = ast.literal_eval(genre_json)  # Convert the JSON text to a list of dictionaries
        return [genre['name'] for genre in genres]  # Extract only the names
    except (SyntaxError, ValueError):
        return []

# Apply the function to the 'genres' column
df['genres'] = df['genres'].apply(extract_genres)

# Display processed data
df.head()
```

<br>

**EXTRACT UNIQUE GENRE LIST**

```python
# Get the list of unique genres in all movies
unique_genres = set(genre for sublist in df['genres'] for genre in sublist)

# Convert it to a sorted list
unique_genres = sorted(unique_genres)

print("Available genres:", unique_genres)
```

```python
# Count the number of unique genres in the dataset
num_genres = len(unique_genres)

print(f"The dataset contains {num_genres} unique genres.")
```

<br>

**DEFINE THE FUNCTION TO FILTER MOVIES BY GENRE**

```python
def get_movies_by_genre(genre):
    """
    Filters movies that contain a specific genre.
    """
    filtered_movies = df[df['genres'].apply(lambda x: genre in x)]
    return filtered_movies[['title']]

# Test with an example
get_movies_by_genre('Crime').head()
```
